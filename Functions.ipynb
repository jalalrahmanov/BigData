### Tasks

```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    answer = sum(1 for char in text if char in vowels)
    return answer
```

```python
print(count_vowels("hello world")) 
```

```python
def second_largest(numbers):
    unique_numbers = list(set(numbers))  
    unique_numbers.sort(reverse=True)   
    return unique_numbers[1]            
```

```python
second_largest([10, 5, 8, 20, 15])
```

```python
def reverse_words(sentence):
    words = sentence.split()     
    reversed_words = [word[::-1] for word in words]
    return " ".join(reversed_words)        
```

```python
reverse_words("hello world")
```

# Filter

```python
def filter_positive(numbers):
    return list(filter(lambda x: x > 0, numbers))
```

```python
filter_positive([-10, 0, 5, -3, 8, 2])
```

```python
def filter_by_letter(words, letter):
    return list(filter(lambda word: word.startswith(letter), words))
```

```python
filter_by_letter(["apple", "banana", "cherry", "apricot", "grape"], "a")
```

# Map

```python
def convert_to_fahrenheit(celsius_temperatures):
    return list(map(lambda c: c * 9/5 + 32, celsius_temperatures))
```

```python
convert_to_fahrenheit([0, 20, 30, 100])
```

```python
def string_lengths(strings):
    return list(map(lambda s: len(s), strings))
```

```python
string_lengths(["apple", "banana", "cherry", "date"])
```
